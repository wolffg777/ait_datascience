<span style="color:red"> MY ANSWER TO: gold_medals=DataFramegolds.groupby(['NOC', 'Year']).Event.apply(lambda x: len(pd.unique(x)))</span>  


This line of code is grouping the columns 'NOC' and 'Year' together into a series with a third column with the number of gold medals per country per year. It does this by reading the number of unique events that the country had a gold in. 

```python
gold_summer = only_summer.query("Medal == 'Gold'") 
gold_summer_medals = gold_summer.groupby(['NOC', 'Year']).Event.apply(lambda x: len(pd.unique(x))) 

list_medals_index = list(gold_summer_medals.index) 
list_medals = list(gold_summer_medals) 

indexdf = pd.DataFrame(list_medals_index, columns = ["NOC", "Year"])
list_medalsdf = pd.DataFrame(list_medals, columns = ['Gold_Medals'])

medaldf = pd.concat([indexdf.reset_index(drop = True), 
                     list_medalsdf.reset_index(drop = True)], axis=1,
    ignore_index=True,).rename(columns = {int("0"): "NOC", int("1"): "Year", int("2"): "Gold_Medals"})

medaldf.head(5)

```

```python
summy = medaldf.groupby(["NOC"])['Gold_Medals'].sum()
medal_sum = pd.DataFrame({"NOC": summy.index, "Gold_Medals": summy.values}) 
                        
medal_sum.head(5)
```

```python
population_data = pd.read_excel('https://math.bme.hu/~pinterj/BevAdat1/Adatok/OlympicsPopulation.xlsx') 

merged = medal_sum.merge(population_data, on = 'NOC') 

merged.head(5)  
```

```python
##average_ratings = movie_mean.apply(lambda x: x.mean())

merged["GM_PC"] = (merged["Gold_Medals"]/merged["POPULATION"])*1000000

merged.head(5)
```

```python
merged = merged.sort_values(by='GM_PC', ascending = False)

merged.head(5)
```

```python
from IPython.core.display import Image, display
from IPython.core.display import HTML

from sklearn import tree 
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
bank_data = pd.read_csv('/Users/nixilis/Desktop/DS_BP/DSHW2/bank2.csv', 
                        delimiter = " ", 
                        names = ['age', 'sex', 'region', 'income', 'married', 'children', 'car','save_acct', 'current_acct', 'mortgage', 'pep']) 

bank_data.head(5)

# numeric_data = bank_data.replace(['NO', 'YES', 'MALE', 'FEMALE'],[0,1,0,1])
# one_hot = pd.get_dummies(numeric_data['region'])
# numeric_data = numeric_data.drop('region', axis = 1)
# numeric_data = numeric_data.join(one_hot)
# ##numeric_data.head() 

# bank_labels = numeric_data['pep']
# bank_attrs  = numeric_data.drop('pep', axis=1) 

# bank_attrs.head() 
```

```python
numeric_data = bank_data.replace(['NO', 'YES', 'MALE', 'FEMALE'],[0,1,0,1])
one_hot = pd.get_dummies(numeric_data['region'])
numeric_data = numeric_data.drop('region', axis = 1)
numeric_data = numeric_data.join(one_hot)
##numeric_data.head() 

bank_labels = numeric_data['pep']
bank_attrs  = numeric_data.drop('pep', axis=1) 

bank_attrs.head() 
```

```python
bank_features_train, bank_features_test, bank_labels_train, bank_labels_test = train_test_split(bank_attrs, bank_labels, test_size=0.33, random_state=42)

dTree = tree.DecisionTreeClassifier(max_depth = int(6))

dTree = dTree.fit(bank_features_train, bank_labels_train) 
```

```python
predictions_binary = dTree.predict(bank_features_test)
predictions_binary

predictions_proba = dTree.predict_proba(bank_features_test)
predictions_proba[0:10] 

cm = metrics.confusion_matrix(bank_labels_test,predictions_binary)

print("Precision: ", metrics.precision_score(bank_labels_test,predictions_binary))
print("Recall: ", metrics.recall_score(bank_labels_test,predictions_binary))
print("Accuracy: ", metrics.accuracy_score(bank_labels_test,predictions_binary))
```

**b)** Now fit the tree using different parameters! Plot the ROC curve of the decision tree obtained in part a) and the new tree in the same figure (with different colors). Also plot the *y=x* diagonal line!

```python
fpr, tpr, threshold = metrics.roc_curve(bank_labels_test,predictions_binary) 
roc_auc = metrics.auc(fpr, tpr) 
```

```python
##new tree 
dTree2 = tree.DecisionTreeClassifier(max_depth = int(10), criterion = "entropy", min_samples_leaf = int(2), min_impurity_decrease = float(0.01))

dTree2 = dTree2.fit(bank_features_train, bank_labels_train) 

predictions_binary2 = dTree2.predict(bank_features_test)
predictions_binary2

predictions_proba2 = dTree2.predict_proba(bank_features_test)
predictions_proba2[0:10] 

cm = metrics.confusion_matrix(bank_labels_test,predictions_binary2)

print("Precision: ", metrics.precision_score(bank_labels_test,predictions_binary2))
print("Recall: ", metrics.recall_score(bank_labels_test,predictions_binary2))
print("Accuracy: ", metrics.accuracy_score(bank_labels_test,predictions_binary2)) 

fpr2, tpr2, threshold = metrics.roc_curve(bank_labels_test,predictions_binary2) 
roc_auc2 = metrics.auc(fpr2, tpr2) 

plt.title('Receiver Operating Characteristic')
plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
plt.plot(fpr2, tpr2, 'b', label = 'AUC = %0.2f' % roc_auc2, color = 'cyan')
plt.legend(loc = 'lower right')
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show() 
```
